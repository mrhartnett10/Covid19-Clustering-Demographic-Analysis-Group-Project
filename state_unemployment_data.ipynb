```python
import pandas as pd
import os
```

```python
#Reading in first state's data so I can have something to append to

data = pd.read_excel('data/unemployment_rate/alabama_2020.xlsx', header = 10)
data = data.drop(columns = 'Year')
data
```

```python
#Getting list of all file names

file_names = [os.listdir('data/unemployment_rate')[x] for x in range(len(os.listdir('data/unemployment_rate')))]
```

```python
#Appending all state data to original alabama dataset

for file in file_names[1 : ]:
    temp = pd.read_excel(f'data/unemployment_rate/{file}', header = 10)
    temp = temp.drop(columns = 'Year')
    data = data.append(temp)
```

```python
#Extracting state names from file names

state_names = [file.split('_')[0] for file in file_names]
```

```python
#Adding the state name

data['state'] = state_names
data
```

```python
data.to_csv('data/unemployment_rate/unemployment_all_states_2020.csv')
```
